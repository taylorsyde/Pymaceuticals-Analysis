```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0, 130, 10)
x_axis
```

```python
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
plt.hlines(0, 0, 120, alpha=1)
plt.ylim(0,100)
plt.xlim(0,125)
plt.grid()
```

```python
plt.plot(x_axis, danger_drop, linewidth=1.5, color="blue", label="Danger Drop")
plt.plot(x_axis, railgun, linewidth=1.5, color="red", label="Railgun")
```

```python
plt.title("Roller Coaster Speeds Over Time")
plt.xlabel("Time in Seconds")
plt.ylabel("Speed of Coaster in MPH")
plt.legend(loc="lower right")
```
