```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_df = pd.read_csv("Resources/got.csv")
got_df.head()
```

```python
# Get attacker and defender data
attackers = got_df["attacker_king"].value_counts()
defenders = got_df["defender_king"].value_counts()
```

```python
# Get total battle data
battle_totals = attackers.add(defenders, fill_value=0)
battle_totals
```

```python
# Configure plot and ticks
battle_totals.plot(kind="bar", facecolor="purple")
```

```python
# Set textual properties
plt.title("The Bloodthirst of King")
plt.xlabel("Kings")
```
