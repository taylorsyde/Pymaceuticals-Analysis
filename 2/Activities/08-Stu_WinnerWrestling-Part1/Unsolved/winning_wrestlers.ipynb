```python
# Import the necessary modules/dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Bring each CSV into a separate data frame
wrestler_data_2013 = pd.read_csv("../Resources/WWE-Data-2013.csv")
wrestler_data_2014 = pd.read_csv("../Resources/WWE-Data-2014.csv")
wrestler_data_2015 = pd.read_csv("../Resources/WWE-Data-2015.csv")
wrestler_data_2016 = pd.read_csv("../Resources/WWE-Data-2016.csv")
```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)
combined_wrestler_data = pd.merge(wrestler_data_2013, wrestler_data_2014, on='Wrestler', how='outer',suffixes=(' in 2013', ' in 2014'))
combined_wrestler_data.head()
```

```python
# Merge our newly combined dataframe with the 2015 dataframe
combined2_wrestler_data = pd.merge(combined_wrestler_data, wrestler_data_2015, on='Wrestler',how='outer')
combined2_wrestler_data.head()
```
